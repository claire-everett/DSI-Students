## Comparing models with different states

```python
import pandas as pd
from glob import glob
import numpy as np
import Conspecific_Tracking_ta_edit_117 as Cs
from Conspecific_Tracking_ta_edit_117 import getfiltereddata
import ssm
from ssm.util import find_permutation
from ssm.plots import gradient_cmap, white_to_color_cmap
import matplotlib.pyplot as plt
import math
```

```python
time_bins = obs.shape[0]
color_names = [
    "windows blue",
    "red",
    "amber",
    "faded green",
    "dusty purple",
    "orange"
    ]
import seaborn as sns
colors = sns.xkcd_palette(color_names)
sns.set_style("white")
sns.set_context("talk")
cmap = gradient_cmap(colors)
```

```python
h5_files = sorted(glob('*.h5'))
data_auto1_filt,data_auto2_filt = getfiltereddata(h5_files)

fish1=data_auto1_filt
fish1=fish1.dropna()

periculum=Cs.auto_scoring_get_opdeg(data_auto1_filt)
peri_array=periculum.array
angle=Cs.gaze_tracking(data_auto1_filt,data_auto2_filt)

obs=np.vstack((periculum,angle)).T
obs=obs.astype(float)
obs=obs[~np.isnan(obs).any(axis=1)]
```

```python
def num_params(model):
    n=model.K
    dim=model.D
    return n-1+n*(n-1)+n*dim+n*(dim*dim+dim)/2

def BIC(model,obs):
    return -2*model.log_likelihood(obs)+num_params(model)*math.log(obs.shape[0])

def AIC(model,obs):
    return -2*model.log_likelihood(obs)+2*num_params(model)
```

```python
N_iters = 50
num_states=2
obs_dim=2
hmm_2state = ssm.HMM(num_states, obs_dim, observations="gaussian")
hmm_2state_ll = hmm_2state.fit(obs, method="em", num_iters=N_iters)
```

```python
hmm_z_2 = hmm_2state.most_likely_states(obs)
plt.imshow(hmm_z_2[None,:], aspect="auto", cmap=cmap, vmin=0, vmax=len(colors)-1)
plt.xlim(0, time_bins)
plt.ylabel("$z_{\\mathrm{inferred}}$")
plt.yticks([])
plt.xlabel("time")
```

```python
model_aic=[]
model_bic=[]
```

```python
model_aic.append(AIC(hmm_2state,obs))
model_bic.append(BIC(hmm_2state,obs))
```

```python
num_states=3
hmm_3state = ssm.HMM(num_states, obs_dim, observations="gaussian")
hmm_3state_ll = hmm_3state.fit(obs, method="em", num_iters=N_iters)
hmm_z_3 = hmm_3state.most_likely_states(obs)
plt.imshow(hmm_z_3[None,:], aspect="auto", cmap=cmap, vmin=0, vmax=len(colors)-1)
plt.xlim(0, time_bins)
plt.ylabel("$z_{\\mathrm{inferred}}$")
plt.yticks([])
plt.xlabel("time")
```

```python
model_aic.append(AIC(hmm_3state,obs))
model_bic.append(BIC(hmm_3state,obs))
```

```python
for i in range(4,9):
    num_states=i
    hmm_state = ssm.HMM(num_states, obs_dim, observations="gaussian")
    hmm_state.fit(obs, method="em", num_iters=N_iters)
    model_aic.append(AIC(hmm_state,obs))
    model_bic.append(BIC(hmm_state,obs))
```

```python
for i in range(9,15):
    num_states=i
    hmm_state = ssm.HMM(num_states, obs_dim, observations="gaussian")
    hmm_state.fit(obs, method="em", num_iters=N_iters)
    model_aic.append(AIC(hmm_state,obs))
    model_bic.append(BIC(hmm_state,obs))
```

```python
print(model_aic)
print(model_bic)
```

```python
dimension=list(range(2,15))
plt.plot(dimension,model_aic)
plt.title("AIC with different num_states")
plt.xlabel("dimension")
plt.ylabel("AIC")
plt.show()
```
